```python
import numpy as np
import matplotlib.pyplot as plt

#mean of each feature
x1 = np.sum(df['Age']) / (len(df) - 1)
x2 = np.sum(df['BMI']) / (len(df) - 1)
x3 = np.sum(df['Glucose']) / (len(df) - 1)
x4 = np.sum(df['Insulin']) / (len(df) - 1)
x5 = np.sum(df['HOMA']) / (len(df) - 1)
x6 = np.sum(df['Leptin']) / (len(df) - 1)
x7 = np.sum(df['Adiponectin']) / (len(df) - 1)
x8 = np.sum(df['Resistin']) / (len(df) - 1)
x9 = np.sum(df['MCP.1']) / (len(df) - 1)

print("Mean of Age:", x1)
print("Mean of BMI:", x2)
print("Mean of Glucose:", x3)
print("Mean of Insulin:", x4)
print("Mean of HOMA:", x5)
print("Mean of Leptin:", x6)
print("Mean of Adiponectin:", x7)
print("Mean of Resistin:", x8)
print("Mean of MCP.1:", x9)

x_mean = [x1, x2, x3, x4, x5, x6, x7, x8, x9]

cov_matrix = np.zeros((9, 9))

#covariance matrix
for i in range(9):
    for j in range(9):
        if i == j:
            cov_matrix[i, j] = np.sum((df.iloc[:, i] - x_mean[i]) ** 2) / (len(df) - 1)
        else:
            cov_matrix[i, j] = np.sum((df.iloc[:, i] - x_mean[i]) * (df.iloc[:, j] - x_mean[j])) / (len(df) - 1)

print("Covariance Matrix:")
print(cov_matrix)

#eigenvalues and eigenvectors of the covariance matrix
eigenvalues,eigenvectors = np.linalg.eig(cov_matrix)

idx = np.argsort(-eigenvalues)
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:")
print(eigenvectors)

k = 3
pc1 = eigenvectors[:, 0]
pc2 = eigenvectors[:, 1]
pc3 = eigenvectors[:, 2]

print("First principal component :", pc1)
print("Second principal component :", pc2)
print("Third principal component :", pc3)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

pc1_scores = df.iloc[:, :9].values @ pc1
pc2_scores = df.iloc[:, :9].values @ pc2
pc3_scores = df.iloc[:, :9].values @ pc3

set1 = df[df['Classification'] == 1]
set2 = df[df['Classification'] == 2]

ax.scatter(pc1_scores[set1.index], pc2_scores[set1.index], pc3_scores[set1.index], color='r', label='Class 1')
ax.scatter(pc1_scores[set2.index], pc2_scores[set2.index], pc3_scores[set2.index], color='g', label='Class 2')

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
plt.title('Data Points in the First Three Principal Components')
plt.legend()
plt.show()
```
